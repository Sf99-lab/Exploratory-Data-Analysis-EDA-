```python
import numpy as np
import matplotlib.pyplot as plt

#plt.style.use('./deeplearning.mplstyle')
plt.style.use('ggplot')
```

```python
# x_train is the input variable (size in sq. feet)
# y_train is the target (price in dollars)

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

***Number of Training Examples***

```python
# m is the number of training example
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
i = 0 # Change this to 1 to see (x^1, x^2)
x_i = x_train[i]
y_i = y_train[i]
print(f"x^({i}), y^({i}) = ({x_i}), ({y_i})")
```

***Plotting the data***

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price (US $)")
plt.show()
```

***Model function***

```python
w=100 # You can try w = 100 to best fit the line
b=100

print(f"w: {w}")
print(f"b: {b}")
```

```python
def calculate_model_output( w, b, x):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
      f_wb[i] = w * x[i] + b
    print(f_wb)
    return f_wb
```

```python
#calculate model ouput
temp_f_wb = calculate_model_output(w, b, x_train)

#plot model prediction
plt.plot(x_train, temp_f_wb, c='b', label='Our Prediction')

#plot the data points
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

plt.title("Housing Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price (US $)")
plt.legend()
plt.show()
```
